```
#Q1
#I have used Euler Discretization
```

```
import math
import numpy as np

def cir_euler(x0, kappa, theta, sigma, T, N, S):
  dt = T/N    
  x = x0
  c=0
  for i in range(0,S):
    x = x0
    for t in range(1,N+1):
      x = x + (kappa * (theta - x)* dt + np.sqrt(x) * sigma * np.random.standard_normal() * math.sqrt(dt))
      print(x, end = ' ')
      c = c+1
      if c%10 == 0:
        print("")     
```

```
cir_euler(0.5,0.3,0.2,0.1,1,20,8)
```

```
def cir_exact(x0, kappa, theta, sigma, T, N, S):
  dt = T / N
  x = x0
  d = 4 * kappa * theta / sigma ** 2
  c = (sigma ** 2 * (1 - math.exp(-kappa * dt))) / (4 * kappa)
  count=0
  if d > 1:
    for t in range(1, N + 1):
      l = x * math.exp(-kappa * dt) / c
      chi = np.random.chisquare(d - 1)
      x = c * ((np.random.standard_normal() + np.sqrt(l)) ** 2 + chi)
      print(x,end=' ')
      count = count + 1
      if count%10==0:
        print("")
  else:
    for t in range(1, N + 1):
      l = x * math.exp(-kappa * dt) / c
      if (l + 2*math.log(np.random.uniform())) <= 0:
        chi = np.random.chisquare(d)
        x = c * chi
        print(x, end = '')
        count = count + 1
        if count%10==0:
          print("")
      else:
        chi = np.random.chisquare(d) + (np.random.standard_normal() + math.sqrt(l + 2*math.log(np.random.uniform())))**2 + np.random.standard_normal()**2
        x = c * chi
        print(x, end = '')
        count = count + 1
        if count%10==0:
          print("")
```

```
cir_exact(0.5,0.3,0.2,0.1,1,20,8)
```
